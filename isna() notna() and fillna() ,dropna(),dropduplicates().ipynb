```python
import pandas as pd
import numpy as np
```

```python
dt={
  'name':['Alfred','Batman','Catwoman','Ironman',np.nan],
    
    'born':[pd.Timestamp("1950-03-21"),pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT,pd.Timestamp("1999-03-02")],
    
    'toy':[np.nan,'Batmobile','bullwhip',np.nan,np.nan],
    
    'rating':[4,5,3,4,np.nan],
    'check':[np.nan,np.nan,np.nan,np.nan,np.nan]
    
}
```

```python
df=pd.DataFrame(dt)
```

```python
df
```

```python
df.dropna()   # row level
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how='all')
```

```python
df.dropna(how='all',axis=1)
```

```python
# Keep only the rows with at least 2 non-Na value

df.dropna(thresh=2)
```

```python
df.dropna(thresh=1)
```

# drop_duplicates()

```python
dt={
  'name':['Alfred','Alfred','Catwoman','Ironman',np.nan],
    
    'born':[pd.Timestamp("1950-03-21"),pd.Timestamp("1950-03-21"),pd.Timestamp("1940-04-25"),pd.NaT,pd.Timestamp("1999-03-02")],
    
    'toy':['Batmobile','Batmobile','bullwhip',np.nan,np.nan],
    
    'rating':[4,4,3,4,np.nan],
    'check':[np.nan,np.nan,np.nan,np.nan,np.nan]
    
}
df=pd.DataFrame(dt)
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True)
```
