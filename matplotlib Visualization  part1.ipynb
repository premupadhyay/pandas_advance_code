```python
x=np.arange(1,20)
y1=np.sqrt(x)
y2=np.log(x)
y3=np.sin(x)
```

```python
plt.figure(figsize=(7,7))
plt.plot(x,y1,color='blue')
plt.plot(x,y2,color='red')
plt.plot(x,y3,color='green')
plt.legend(labels=["Square root",'log','sign'])
plt.show()
```

# Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
classes=['Python','R','AI','Java']
c1=[30,10,20,40]
c2=[20,25,30,70]
c3=[45,23,12,10]
c4=[5,3,2,1]
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(8,4))
plt.bar(classes,c1,width=0.2,align='edge',color='b',edgecolor='g',
        linewidth=2,alpha=0.5,linestyle='--',label="Class1") #visible=False
```

```python
plt.barh(classes,c1,height=0.2,align='edge',color='r',edgecolor='g',linewidth=2,
         alpha=0.5,linestyle='--',label="Class1")
```

```python
import numpy as np
data1=[23,85,72,43,52]
data2=[42,35,21,16,9]
data3=[43,77,99,88,121]
width=0.3
plt.bar(np.arange(len(data1)),data1,width=width,align='edge',color='r',edgecolor='g',linewidth=2,
         alpha=0.5,linestyle='--',)

plt.bar(np.arange(len(data3)),data1,width=width,align='edge',color='g',edgecolor='g',linewidth=2,
         alpha=0.5,linestyle='--',bottom=data1)

plt.bar(np.arange(len(data2)),data2,width=width,align='edge',color='b',edgecolor='r',linewidth=2,
         alpha=0.5,linestyle='--',bottom=data2)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("gplay1.csv")
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
df.drop(index=[5542],inplace=True)
df.fillna(df['Rating'].mean(),inplace=True)
df['Size'].replace(['k','M',],'',regex=True,inplace=True)
df['Size'].replace('Varies with device',0,regex=True,inplace=True)

df['Size'] = df['Size'].astype(float)
df['Installs'].replace([',','\+'],'',regex=True,inplace=True)
df['Installs'] = df['Installs'].astype(int)
```

```python
x=list(df['Rating'])
y=list(df['Reviews'])
```

```python
plt.scatter(x,y)
plt.title("google play store scatter plot")
plt.xlabel("Rating data")
plt.ylabel("Review data")
```

```python
x1=list(map(lambda x: np.sqrt(x),x))
y1=list(map(lambda x: np.log(y),y))
```

```python
len(x1)
```

```python
y1
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y,c='b',marker="+",s=60,alpha=0.2,linewidth=2,label="Rating")
plt.scatter(x1,y1,c='g',marker="+",s=60,alpha=0.2,linewidth=2,label="Rating")
plt.title("google play store scatter plot")
plt.xlabel("Rating data")
plt.ylabel("Review data")
plt.legend()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['Rating'],df['Price'],color='red')
plt.title("Rating vs Price scatter")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['Rating'],df['Price'],color='red')
plt.scatter(df['Rating']*2,df['Price']*2,color='green')
plt.scatter(df['Rating']/2,df['Price']/2,color='blue')
plt.title("Rating vs Price scatter")
plt.xlabel("Price")
plt.ylabel("Ridth")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['Rating'],df['Price'],color='red',label="first")
plt.scatter(df['Rating']+3000,df['Price']+500,color='green',label='second')
plt.scatter(df['Rating']*6,df['Price']/2,color='blue',label='third')
plt.plot(df.index,df['Rating']*2,color='red',label='line')
plt.title("Rating vs Price scatter")
#plt.legend(loc='upper right')
#plt.legend(loc='lower right')
plt.legend(ncol=4,labelcolor=['blue','red','green','yellow'],loc='upper right',fontsize=12)
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

# pai chart

```python
import matplotlib.pyplot as plt
classes=['Python','R','AI','Java']
c1=[30,10,20,40]
explode=[0,0,0.2,0]
color=['y','b','r','g']
textprops={"fontsize":15}
plt.pie(c1,labels=classes,shadow=True,startangle=180,autopct="%0.1f%%",explode=explode,colors=color,radius=2,textprops=textprops)
plt.show()
```
