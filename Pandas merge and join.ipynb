```python
import pandas as pd
left = pd.DataFrame({
         'id':[1,2,3,4,5],
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
         {'id':[1,2,3,4,5],
         'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
         'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on='id')   # by default use inner join 
```

```python
pd.merge(left,right,on='subject_id') # by default use inner join 
```

```python
pd.merge(left,right,how='left',on='subject_id') # left join
```

```python
pd.merge(left,right,how='right',on='subject_id') #  right join 
```

```python
# Merge based on more than on columns
pd.merge(left,right,on=['id','subject_id'])  # inner join
```

```python
pd.merge(left,right,how='left',on=['id','subject_id'])
```

```python
pd.merge(left,right,on=['id','subject_id'],how='outer')
```

```python
pd.merge(left,right,on=['id','subject_id'],how='inner')
```
