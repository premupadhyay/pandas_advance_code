# Box plot

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.rand(10,5),columns=['A','B','C','D','E'])
df.plot.box()
```

```python
df.plot.area()
```

```python
df.plot.scatter(x='C',y='B')
```

```python
df1=df['A']
explode = (0.06, 0.6, 0.05,0,0,0,0,0,0,0)
df1.plot.pie(subplots=True,autopct='%1.0f%%',explode=explode)
```
