```python
url = 'https://www.selenium.dev/blog/2022/'
r = requests.get(url)
soup = BeautifulSoup(r.content)

data = soup.find('div',attrs={'class':'taxonomy taxonomy-terms-cloud taxo-tags'})
for tags in data.findAll('li'):
    print(tags.span.text)
```

```python

import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/Barack_Obama'
r = requests.get(url)
soup = BeautifulSoup(r.content)

data = soup.find('div',attrs={"id":"mw-content-text"})

for td in data.find_all('p'):
    print(td.text)
            
```
