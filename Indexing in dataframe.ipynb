```python
import pandas as pd
ls=[['Jhon','DS',95,'A'],['raw','ML',96,'B'],['Test','DL',70,'C'],['Bob','CS',50,'D'],['Ram','DS',75,'A'],['Rahul','ML',80,'A'],['Test','DL',90,'A'],['Demo','CS',90,'A']]
df=pd.DataFrame(data=ls,columns=['Name','Subject','marks','Grade'])
```

```python
df.loc[~(df['Name']=='Test') & (df['marks']>75)]
```

```python
# 1. Display all record having A grade in ML
df.loc[(df['Subject']=='ML') & (df['Grade']=='A')]

# 2. Display all the record having C and D grade and its not ML and DL
df.loc[(df['Grade']=='C')|(df['Grade']=='D')& ~(df['Subject']=='ML')&~(df['Subject']=='DS')]

# 3. Display only name have DL subject
 ## 1
df[['Name','Subject']].loc[df['Subject']=='DL']


# 4. Display only name having DL or ML subject but A grade 


#df.loc[((df['Subject']=='ML')& (df['Subject']=='ML')) & (df['Grade']=='A' )] ##4


# 5. Display A grade info only Name and Marks
df[['Name', 'marks']].loc[(df['Grade'] == 'A')]
```

```python
import pandas as pd
ls=[['Jhon','DS',95,'A'],['raw','ML',96,'B'],['Test','DL',70,'C'],['Bob','CS',50,'D'],['Ram','DS',75,'A'],['Rahul','ML',80,'A'],['Test','DL',90,'A'],['Demo','CS',90,'A']]
df=pd.DataFrame(data=ls,columns=['Name','Subject','marks','Grade'])

#-----------Assigment-2-------------------
# 1. Display all record having A grade in ML
# 2. Display all the record having C and D grade and its not ML and DL
# 3. Display only name have DL subject
# 4. Display only name having DL or ML subject but A grade 
# 5. Display A grade info only Name and Marks
# 6. if grade is D add 5 marks to marks columns
# 7. if name is test and subject is DL change the grade 'A+'
# 8. create new columns percentage and add percentage of all student 
```

```python
#Q1. if grade is D add 5 marks to marks columns
df.loc[df['Grade']=='D','marks']=df['marks']+5


#Q2. if name is test and subject is DL change the grade 'A+'
#df.loc[(df['Name']=='Test')&(df['Subject']=='DL'),'Grade']="A+"

# #Q3. create new columns percentage and add percentage of all student 

df['percentage']=df['marks'].astype(str)+"%"
```
