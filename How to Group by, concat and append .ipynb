```python
import pandas as pd
import numpy as np
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'Kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
```

```python
df.groupby('Year').count()
```

```python
for data in gb:
    print(data)
```

```python
gb1=df.groupby(['Year','Team'])
```

```python
for data in gb1:
    print(data)
```

# Agg function in group by

```python
df.groupby('Rank')['Team'].sum()
```

```python
df.groupby('Rank')['Team'].count()
```

```python
df.groupby('Year')['Points'].mean()
```

```python
df.groupby('Year')['Points'].agg([np.sum,np.mean,np.std,np.size])
```

# Filtration 

```python
gb=df.groupby('Year')
```

```python
gb.apply(lambda x: x[x['Rank']>3])
```

```python
gb.apply(lambda x: x[(x['Rank']==4) & (x['Year']==2015)])
```

```python
gb.apply(lambda x: x[~(x['Rank']==4) & (x['Year']==2015)])
```
